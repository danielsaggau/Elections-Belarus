# Creating the data set
### Loading the libraries

```python
import pandas as pd
import numpy as np
```

### Here we load all the needed data set to merge them together

```python
votes=pd.read_excel('belarus_elections.xlsx')
commission=pd.read_csv('commission.csv')
parent=pd.read_csv('member-00_parent.csv')
qgis=pd.read_csv('belarus-qgis.csv')
```

### Join votes with commision by the code of commission

```python
df=votes.join(commission.set_index('code'), on='Код комиссии')
df.columns
```

### Changing the type of values in qgis['parent_id']

```python
qgis['parent_id'] = qgis['parent_id'].map(lambda x: str(x).rstrip('.')).astype(float)
qgis['parent_id'].unique()
```

### Creating a table with commissions features

```python
df_merge = pd.merge(df, commission_features, left_on='parent_id', right_on='commission_id')
df_merge.columns
```
